### Theoritical rules for dfs:
1. Start at any random node.
2. start visiting the univisted neighbours, once you pick a neighbour then you go forward then traverse the next node of the neighbour, you will not comeback unless all nodes are visied from a particular node.
3. Start and end of the node will be same, this shows the dfs is

Example:
0 -> 1

1 -> 3 (you wont go back to 0)

3 -> 2 (from 2 there is nothing to visit, because (3,0) are already visited, so you step back)

3 -> 4 (from 4 there is nothing to visit, so you step back to 1 and from 1 nothing to visit so step back to 0) --> your dfs is done.
